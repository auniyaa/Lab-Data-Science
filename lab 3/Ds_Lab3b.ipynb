```r
# Input vector
scores <- c(33,24,54,94,16,89,60,6,77,61,13,44,26,24,73,73,90,39,90,54)

pass<- scores > 49
print(pass)

grade<- c()

for(i in scores){
    if(i>=90 & i<=100 ){
        grade <-c(grade,"A")
    }else if (i>=80 & i<= 89 ){
        grade <-c(grade,"B")
    }else if(i>= 70 & i <=79){
        grade<-c(grade,"C")
    }else if (i>=60 & i <=69){
        grade<-c(grade,"D")
    }else if (i>=50 & i <=59){
        grade<-c(grade,"E")
    }else{
        grade<-c(grade,"F")
    }
  }

table(grade)



```

```r
# Create a list containing a vector and a list.
list_data <- list(c("Robert","Hemsworth","Scarlett","Evans","Pratt","Larson","Paul","Simu","Renner"),
                     list(c(59 ,71 ,83 ,68 ,65 ,57 ,62 ,92 ,92)))
# Give names to the elements in the list.
names(list_data) <- c("Student Name", "Exam Score")
# Show the list.
print(list_data)

scores <- unlist(list_data$'Exam Score')
student <- unlist(list_data$'Student Name')

highest<- max(scores)
lowest<- min(scores)
average<- mean(scores)
top_student <-student[scores==highest]
lowest_student <-student[scores==lowest]

cat("Highest Score:",highest,"\n")
cat("Lowest Score:",lowest,"\n")
cat("Average Score:",average,"\n")
cat("Student with highest score:",top_student,"\n")
cat("Student with lowest score:",lowest_student,"\n")

```

```r
list_data <- list(
  `Student ` = c("Robert","Hemsworth","Scarlett","Evans","Pratt","Larson","Holland","Paul","Simu","Renner"),
  Chemistry = c(59,71,83,68,65,57,62,92,92,59),
  Physics = c(89,86,65,52,60,67,40,77,90,61)
)

# Show the list
print(list_data)

chemScore <- list_data$Chemistry
physicScore <- list_data$Physics
student <- list_data$`Student `

fail_chem<-sum(chemScore<=49)
fail_phys<-sum(physicScore<=49)

highChem<- max(chemScore)
top_chem_student <- student[chemScore == highChem]

highest_phys <- max(physicScore)
top_phys_student <- student[physicScore == highest_phys]

cat("Number of students who failed Chemistry (<=49):", fail_chem, "\n")
cat("Number of students who failed Physics (<=49):", fail_phys, "\n")
cat("Highest Chemistry score:", highChem ,"- Student(s):", top_chem_student, "\n")
cat("Highest Physics score:", highest_phys, "- Student(s):", top_phys_student, "\n")
```
