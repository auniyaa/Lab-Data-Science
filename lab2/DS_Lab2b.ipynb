```r
x=13

if (x <= 15) {
  cat("UNDERWEIGHT: TRUE\n")
} else {
  cat("UNDERWEIGHT: FALSE\n")
}
if  (x >= 15 && x <= 24.9) {
  cat("NORMAL: TRUE\n")
} else {
  cat("NORMAL: FALSE\n")
}
if  (x >= 25 && x <= 39.9) {
  cat("OVERWEIGHT: TRUE\n")
} else {
  cat("OVERWEIGHT: FALSE\n")
}
if  (x >= 40.0) {
  cat("OBESE: TRUE\n")
} else {
  cat("OBESE: FALSE\n")
}
```
